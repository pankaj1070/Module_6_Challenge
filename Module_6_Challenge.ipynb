```python
# Import the required libraries and dependencies
import pandas as pd
import hvplot.pandas
from pathlib import Path
```

```python
# Using the read_csv function and Path module, create a DataFrame 
# by importing the sfo_neighborhoods_census_data.csv file from the data source folder
sfo_data_df = pd.read_csv('C:/Users/panka/Documents/Homework/Module_6_Challenge/sfo_neighborhoods_census_data.csv')

# Review the first and last five rows of the DataFrame
sfo_data_df
```

```python
# Create a numerical aggregation that groups the data by the year and then averages the results.
housing_units_by_year = sfo_data_df.groupby("year").mean()

# Review the DataFrame
housing_units_by_year
```

```python
# Use the hvplot function to plot the housing_units_by_year DataFrame as a bar chart 
# Make the x-axis represent the year and the y-axis represent the housing_units
# Style and format the line plot to ensure a professionally styled visualisation

housing_units_by_year_plot = housing_units_by_year.hvplot.bar(x="year",y="housing_units",ylim=(365000, 385000),label="Housing units in San francisco from 2010 to 2016").opts(yformatter='%.0f').opts(xlabel='Year',ylabel='Housing Units')
housing_units_by_year_plot
```

```python
# What’s the overall trend in housing units over the period that you’re analysing?

# Ans - The overall trend in San Francisco shows that there is a constant increase in number of housing units from 2010 to 2016.
# The Housing units increased from 372560 units in 2010 to 384242 units in 2016.
```

```python
# Calculate and Plot the Average Sale Prices per Square Foot

# Create a numerical aggregation by grouping the data by year and averaging the results
gross_rent = sfo_data_df.groupby("year").mean().sort_values("gross_rent",ascending=True)

# Review the resulting DataFrame
gross_rent
```

```python
# What is the lowest gross rent reported for the years included in the DataFrame?

# Answer : The lowest gross rent reported for the years include in above dataframe is $1239.
```

```python
# Filter out the housing_units column, creating a new DataFrame 
# Keep only sale_price_sqr_foot and gross_rent averages per year
prices_square_foot_by_year = gross_rent[['sale_price_sqr_foot','gross_rent']]

# Review the DataFrame
prices_square_foot_by_year
```

```python
# Use hvPlot to plot the prices_square_foot_by_year DataFrame as a line plot

# Style and format the line plot to ensure a professionally styled visualisation

prices_square_foot_by_year_plot = prices_square_foot_by_year.hvplot.line(x="year",y=["sale_price_sqr_foot","gross_rent"],label="Sale Price Per Square Foot and Average Gross Rent - 2010-2016 - San Francisco").opts(yformatter='%.0f',xlabel='Year',ylabel='Gross Rent/Sale Price Per Square Foot')
prices_square_foot_by_year_plot
```

```python
# Did any year experience a drop in the average sale price per square foot compared to the previous year?

# Ans - The average sale price dercreased only between the year 2010 to 2011. The average price per sq foot decreased from 
# $369.34/sqfoot to $341.9/sqfoot

# If so, did the gross rent increase or decrease during that year?

# Ans: Even with the decreasing prices, the gross rent increased from 1239 to 1530 between 2010-2011. 
```

```python
# Load neighborhoods coordinates data
neighborhood_locations_df = pd.read_csv(
    Path("C:/Users/panka/Documents/Homework/Module_6_Challenge/neighborhoods_coordinates.csv"), 
    index_col="Neighborhood", 
    low_memory=False
)

# Review the DataFrame
neighborhood_locations_df
```

```python
# Calculate the mean values for each neighborhood using sfo_data_df dataframe
all_neighborhood_info_df = sfo_data_df.groupby("neighborhood").mean()

# Round the numbers in dataframe to 2 decimal points 
all_neighborhood_info_df = round(all_neighborhood_info_df,2)

# Review the resulting DataFrame
all_neighborhood_info_df
```

```python
# Using the Pandas `concat` function, join the 
# neighborhood_locations_df and the all_neighborhood_info_df DataFrame
# The axis of the concatenation is "columns".
# The concat function will automatially combine columns with
# identical information, while keeping the additional columns.
all_neighborhoods_df = pd.concat(
    [neighborhood_locations_df, all_neighborhood_info_df], 
    axis="columns",
    sort=False
)

# Review the resulting DataFrame
all_neighborhoods_df
```

```python
# Call the dropna function to remove any neighborhoods that do not have data
all_neighborhoods_df = all_neighborhoods_df.reset_index().dropna()
```

```python
# Rename the "index" column as "Neighborhood" for use in the Visualization
all_neighborhoods_df = all_neighborhoods_df.rename(columns={"index": "Neighborhood"})
```

```python
# Review the resulting DataFrame
all_neighborhoods_df.head()
```

```python
# Create a plot to analyze neighborhood info
# Set the geo parameter to True
# Set the size parameter to “sale_price_sqr_foot”
# Set the color parameter to “gross_rent
# Show all columns to help answer all the questions
# Use OSM for open street Map

all_neighborhoods_df.hvplot.points('Lon', 'Lat', geo=True, size='sale_price_sqr_foot',color='gross_rent', hover_cols='all',
                                   frame_width=700,frame_height=500,tiles='OSM',label='All Neighborhood locations by Average Rent').opts(yformatter='%.0f')
```
